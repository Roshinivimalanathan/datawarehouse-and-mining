## c) Data Transformation and Normalization (using wine data set)

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from pandas import read_csv
from pandas import DataFrame
from pandas.plotting import scatter_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Normalizer

df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",sep=";")
df
```

```python
df.isnull().sum()
```

```python
def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final
```

```python
df_outlier_removed=remove_outlier_IQR(df.quality)
df_outlier_removed=pd.DataFrame(df_outlier_removed)
ind_diff=df.index.difference(df_outlier_removed.index)

for i in range(0, len(ind_diff),1):
    df_final=df.drop([ind_diff[i]])
    df=df_final
```

```python
len(ind_diff)
```

```python
df.drop_duplicates(inplace=True)
df
```

### Transformations

```python

# LOG TRANSFORMATION
np.random.seed(0)
data = np.random.beta(a=4, b=15, size=300)

data_log = np.log(data)
data_log

fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].hist(data, edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

axs[0].set_title('Original Data')
axs[1].set_title('Log-Transformed Data')
```

```python
# SQUARE TRANSFORMATION
np.random.seed(0)
data = np.random.beta(a=4, b=15, size=300)

data_log = np.sqrt(data)
data_log

fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].hist(data, edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

axs[0].set_title('Original Data')
axs[1].set_title('Square-Transformed Data')
```

```python
# CUBIC TRANSFORMATION
np.random.seed(0)
data = np.random.beta(a=4, b=15, size=300)

data_log = np.cbrt(data)
data_log

fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].hist(data, edgecolor='black')
axs[1].hist(data_log, edgecolor='black')

axs[0].set_title('Original Data')
axs[1].set_title('Cubic-Transformed Data')
```
